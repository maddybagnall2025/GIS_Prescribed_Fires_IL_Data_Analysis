```python
"""
GIS Data Analysis (Python Script)

Export 300gpi

"""
import numpy as np
import pandas as pd
from plotly.subplots import make_subplots
import plotly.graph_objects as go
import os
```

```python
#https://stackoverflow.com/questions/55698429/different-box-plot-series-traces-within-plotly-subplots

areaVisual = pd.read_csv("PrescribedBurnsData.csv")

range1 = list(range(2000,2006))
range2 = list(range(2006,2011))
range3 = list(range(2011,2016))
range4 = list(range(2016,2021))
range5 = list(range(2021,2023))

endDate = list(areaVisual['SznEndDate'])
UAorNonUA = list(areaVisual['UA'])
areas = list(areaVisual['Acreage'])

UArange1 = []
NonUArange1 = []

UArange2 = []
NonUArange2 = []

UArange3 = []
NonUArange3 = []

UArange4 = []
NonUArange4 = []

UArange5 = []
NonUArange5 = []

for i in range(len(endDate)):
    if any(str(val) in str(endDate[i]) for val in range1):
        if "notUA" in UAorNonUA[i]:
            NonUArange1.append(areas[i])
        else: 
            UArange1.append(areas[i])
    
    elif any(str(val) in str(endDate[i]) for val in range2):
        if "notUA" in UAorNonUA[i]:
            NonUArange2.append(areas[i])
        else: 
            UArange2.append(areas[i])
        
    elif any(str(val) in str(endDate[i]) for val in range3):
        if "notUA" in UAorNonUA[i]:
            NonUArange3.append(areas[i])
        else: 
            UArange3.append(areas[i])
        
    elif any(str(val) in str(endDate[i]) for val in range4):
        if "notUA" in UAorNonUA[i]:
            NonUArange4.append(areas[i])
        else: 
            UArange4.append(areas[i])
    else:
        if "notUA" in UAorNonUA[i]:
            NonUArange5.append(areas[i])
        else: 
            UArange5.append(areas[i])
```

```python
#Graphing Average Acreage of fires in UA and NonUA areas in the five year brackets in boxplots

#X-axes (labels) for boxplots
x1 = ["UA" for val in UArange1]
x1non = ["nonUA" for val in NonUArange1]
x2 = ["UA" for val in UArange2]
x2non = ["nonUA" for val in NonUArange2]
x3 = ["UA" for val in UArange3]
x3non = ["nonUA" for val in NonUArange3]
x4 = ["UA" for val in UArange4]
x4non = ["nonUA" for val in NonUArange4]
x5 = ["UA" for val in UArange5]
x5non = ["nonUA" for val in NonUArange5]

graphingList = [[x1non,NonUArange1], [x1,UArange1], [x2non,NonUArange2], [x2,UArange2], \
                [x3non,NonUArange3], [x3,UArange3], [x4non,NonUArange4], [x4,UArange4],\
                [x5non,NonUArange5], [x5,UArange5]]

#Will contain boxplot traces
traceDict = {}

#Establishes variables
count1 = 2000
count2 = 2005
val = 0
naming = ""
col = 1
colorNum1 = 255
colorNum2 = 0

fig0 = make_subplots(rows=1, cols=5)

#Creates, stores, and appends traces of boxplots
for item in graphingList:
    
    if 'non' in item[0]:
        naming = 'Non-Urban Acreage ' + str(count1) + "-" + str(count2)
    else:
        naming = 'Urban Acreage ' + str(count1) + "-" + str(count2)
   
    traceDict[tuple(item[0])] = go.Violin(
    y = item[1],
    x = item[0],
    name = naming,
    marker=dict(
        color='rgb(' + str(colorNum2) + ',0,' + str(colorNum1) + ')'
        )
    )
    
    fig0.append_trace(traceDict[tuple(item[0])], 1, col)
    
    val = val+1
        
    if val>1:
        count1 = count1+5
        count2 = count2+5
        col = col + 1
        val=0
    
    colorNum1 = colorNum1 - 25
    colorNum2 = colorNum2 + 25
    

#Removes legend
fig0.update_traces(showlegend=False)

#Displays
fig0.show()
```

```python
#Fire Distance Analysis
visual1 = pd.read_csv("FireDistances.csv")

#Creates variables needed
colNums = list(range(2000,2023))
colNames = [str(x) for x in colNums]
#Dictionary with data separated by year
valDict = {}
traceDict = {}

#Populates valDict with data from each column in the dataframe (labeled by year)
for item in colNames:
    visual1[item] = visual1[item].apply(float)
    almostVals = list(visual1[item])
    valDict["{year}".format(year = item)]=[x for x in almostVals if np.isnan(x)==False]

#Sets up subplots
fig1 = make_subplots(rows=1, cols=1)

#Populates traceDict with Violin traces for each year to be appended to the figure
for item in valDict:
    traceDict[item] = go.Box(
    y=valDict[item],
    x=[str(item) for val in valDict[item]],
    name='Distance to Urban Areas in '+str(item),
    marker=dict(
        color='#BE1E34'
        )
    )
    fig1.append_trace(traceDict[item],1,1)

#Sets the figure size 
fig1.update_layout(
    autosize=False,
    width=1000,
    height=500,
    margin=dict(
        l=50,
        r=50,
        b=100,
        t=100,
        pad=4
    )
)

#Removes legend
fig1.update_traces(showlegend=False)

#Displays
fig1.show()


```
